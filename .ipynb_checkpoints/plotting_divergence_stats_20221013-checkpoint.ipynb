# 1. Plotting the histogram of scaffold sizes

```python
%%bash
ls
#head scaffolds_length.txt
```

```python
sizesfile=open('./scaffolds_length.txt','r')

sizes=[int(lines.replace('\n','')) for lines in sizesfile]
print(sizes[0:5])
```

```python
fig,ax=plt.subplots(figsize=(10,5), facecolor='w')

ax.set_xlim(0,2000000)

(n, bins, patches) = ax.hist([x for x in sizes if x > 10000], bins=200, color='#ffc669')
xticks=['{:.1e}'.format(x) for x in range(int(bins[0]),int(bins[-1]),900000)]
ax.set_xticks([x for x in range(int(bins[0]),int(bins[-1]),900000)],xticks)
ax.set_xlabel('Contigs > 10k length')
ax.set_ylabel('Frequency')

ax1 = inset_axes(ax, width="80%", height="70%", loc=1, borderpad=1)
(n1, bins1, patches1) = ax1.hist([x for x in sizes if x > 5000], bins=200, color='#ff9986')
xticks1=['{:.1e}'.format(x) for x in range(int(bins1[0]),int(bins1[-1]),900000)]
ax1.set_xticks([x for x in range(int(bins1[0]),int(bins1[-1]),900000)],xticks1)
ax1.set_xlabel('Contigs > 5k length')

ax2 = inset_axes(ax1, width="80%", height="70%", loc=1, borderpad=1)
(n2, bins2, patches2) = ax2.hist(sizes, bins=200, color='#bb76c4')
xticks2=['{:.1e}'.format(x) for x in range(int(bins2[0]),int(bins2[-1]),900000)]
ax2.set_xticks([x for x in range(int(bins2[0]),int(bins2[-1]),900000)],xticks2)
ax2.set_xlabel('All contigs')

for axx in ax,ax1,ax2:
    [axx.spines[loc].set_visible(False) for loc in ['top','right']]
    plt.tick_params(axis='both',which='both',top=False,right=False,labelbottom=False)

plt.show()
```

# 2. Nucleotide diversity

## 2.1. Window 10000

```python
%%bash
head MFTall.filtTRUE.snps.10000pi.windowed.pi
```

```python
pi10k=pd.read_csv('./MFTall.filtTRUE.snps.10000pi.windowed.pi',sep='\t')
pi10k.head()
```

```python
%%bash
grep 'NW_012157898.1' scaffolds.txt
```

```python
fig,ax=plt.subplots(figsize=(30,4), facecolor='w')

spinescol='#999999'
cmap = mpl.cm.get_cmap('viridis_r')
norm = mpl.colors.Normalize(vmin=pi10k['PI'].min(), vmax=pi10k['PI'].max())

ax.set_xlim(0,max(pi10k.index))
ax.scatter(pi10k.index,pi10k['PI'], color=[cmap(norm(x)) for x in pi10k['PI']], marker='o', s=10)

plt.yticks(fontsize = 16, color=spinescol); plt.xticks(fontsize = 16, color=spinescol)
plt.grid(axis='y')

plt.xlabel('Position index for contigs > 10k',fontsize=17)
plt.ylabel('PI (10k bins)',fontsize=17)

plt.show()
```
